```python
# Use a pipeline as a high-level helper
from transformers import pipeline

pipe = pipeline("audio-classification", model="ehcalabres/wav2vec2-lg-xlsr-en-speech-emotion-recognition")
```

```python
file_path = "/content/myrecording copy.wav"
audio_input, sample_rate = librosa.load(file_path, sr=16000)
```

```python
results_2 = pipe(audio_input)
```

```python
highest = max(results_2, key=lambda x: x['score'])

# Extract the highest label and its score
highest_label = highest['label']
highest_score = highest['score']

# Print the results
print(f"Highest Label: {highest_label}")
print(f"Highest Score: {highest_score:.2f}")
```
